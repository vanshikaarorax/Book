```python
import pandas as pd

books = pd.read_csv("books_with_categories.csv")
#Loads your pre-processed dataset with simplified categories.
```

```python
from transformers import pipeline
classifier = pipeline("text-classification",
                      model="j-hartmann/emotion-english-distilroberta-base",
                      top_k = None,
                      framework="pt",
                      device = "mps")
classifier("I love this!")
#Loads a RoBERTa model fine-tuned for emotion classification using Hugging Face's pipeline.

#top_k=None: return all emotion scores per input, not just top-1.

#device="mps": runs on your Mac's GPU (Metal Performance Shader).

```

```python
import numpy as np

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}
#Prepares empty containers for:

#isbn: list of book ids.

#emotion_scores: dictionary for each emotion label → list of values.
def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}
    for prediction in predictions:
        sorted_predictions = sorted(prediction, key=lambda x: x["label"])
        for index, label in enumerate(emotion_labels):
            per_emotion_scores[label].append(sorted_predictions[index]["score"])
    return {label: np.max(scores) for label, scores in per_emotion_scores.items()}
#For each sentence, record emotion scores.

# Then, for each emotion, take the maximum score across all sentences.

# Returns a dict of {"joy": 0.82, "fear": 0.34, ...}.

# 📌 Why max? So if even one sentence is very emotional, it reflects in the book-level score.

```

```python
for i in range(10):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
        #Runs the emotion classifier for the first 10 book descriptions.
#Appends each book's emotion scores to emotion_scores.
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}

for i in tqdm(range(len(books))):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
        #Same as earlier, but:

#Loops over all books.

# tqdm shows a progress bar.

# Builds the full emotion score dictionary.t
```

```python
emotions_df = pd.DataFrame(emotion_scores)
emotions_df["isbn13"] = isbn
#Creates a new DataFrame with:

#One row per book
```

```python
emotions_df
```

```python
books = pd.merge(books, emotions_df, on = "isbn13")
#Saves the final enriched dataset with emotion scores added.
```
