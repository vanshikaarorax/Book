```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
from dotenv import load_dotenv

load_dotenv()
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
books
```

```python
books["tagged_description"]
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
raw_documents = TextLoader("tagged_description.txt").load()#Loads each line from the .txt file into LangChain as a Document object.
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")#Splits the loaded text into chunks for embedding.
documents = text_splitter.split_documents(raw_documents)#the above chunk size size 0 and seperater \n meaning every line becomes sep document
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=embedding)
#Useing OpenAI Embeddings to turn each book description into a vector.

#Stores all these vectors in Chroma -vector databasse
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]#This line extracts the ISBN and finds the matching row from the original books dataframe.
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)# 50 most similar books

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]# returning matching books from database 
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
